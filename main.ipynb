```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = requests.get('https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=83c03a47-be01-400d-8b1e-58b35c005b1c').text
```

```python
soup = BeautifulSoup(url, 'lxml')

main_div= soup.find_all('div', class_ = 'tUxRFH')

data = {'Product Name':[], 
       'Rating & Review':[], 
       'Rating No':[],
       'Actual Price':[], 
       'Selling Price':[], 
       'Offer':[], 
       'Memory':[], 
       'Display':[], 
       'Camera':[], 
       'Battery':[]
      }
```

```python
for desc in main_div:
    desc_body = desc.find('div', class_='yKfJKb row')
    
    
    mobile_name = desc_body.find('div', class_='KzDlHZ').text
    data['Product Name'].append(mobile_name)
    
    
    rating_review = desc_body.find('span', class_='Wphh3N').text
    data['Rating & Review'].append(rating_review)
    
    rating_no = desc_body.find('div', class_='XQDdHH').text
    data['Rating No'].append(rating_no)
    
    actual_price = desc_body.find('div', class_='yRaY8j ZYYwLA').text
    data['Actual Price'].append(actual_price)
    
    selling_price = desc_body.find('div', class_='Nx9bqj _4b5DiR').text
    data['Selling Price'].append(selling_price)
    
    
    offer = desc_body.find('div', class_='UkUFwK').text
    if len(offer) == 0:
        data['Offer'].append('No Offer')
    else:    
        data['Offer'].append(offer)
    
    
    specification = desc_body.find('div', class_='_6NESgJ')
    
    data['Memory'].append(specification.find_all('li')[0].text)
    data['Display'].append(specification.find_all('li')[1].text)   
    data['Camera'].append(specification.find_all('li')[2].text)   
    data['Battery'].append(specification.find_all('li')[3].text)


```

```python
pd.DataFrame(data)
```

```python
final_data = pd.DataFrame(data)
final_data.to_csv('Flipkart_Data.csv', index= False)
```

```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")

db = client['Flipkart_Data']
collection = db['mobile_phones']
```

```python
final_dict = final_data.to_dict('records')
collection.insert_many(final_dict)
```
